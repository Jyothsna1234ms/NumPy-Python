```python
x= np.arange(1,11)
y = 3*x+5
plt.title("Matplotlib")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave from")

plt.plot(x,y)
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)   #2 rows, 1 column 

plt.plot(x,y_sin, 'r--')
plt.title('Sine Graph')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine Graph')

```

```python
#Box Plot

data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data, vert=True,patch_artist=True)        #circles are outliers   patch_artist-color, vert is false- rotates it
```

```python
labels= 'Pytho,', 'C++', 'Ruby', 'Java'

sizes = [215,130,45,210]
colors=['gold', 'yellowgreen', 'lightcoral', 'skyblue']
explode=(0.1,0,0,0)   #slices distance

plt.pie(sizes, explode=explode,labels=labels,colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('Equal')
```
